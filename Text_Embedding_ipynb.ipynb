## Finding Similarities Using Embedding

```python
import pandas as pd
import numpy as np
```

```python
df  = pd.read_csv('words.csv')
df
```

```python
!pip install tiktoken -q
import tiktoken
```

```python
#uses the tiktoken library to get the tokenizer (encoder) that matches the OpenAI model text-embedding-ada-002.
enc = tiktoken.encoding_for_model('text-embedding-ada-002')
print(len(enc.encode('above us only sky')))
```

```python
words = list(df['text'])
print(words)
```

```python
#calculate total tokens in "word.csv"
total_tokens = sum([len(enc.encode(word)) for word in words])
print(f'Total tokens: {total_tokens}')
```

```python
cost = total_tokens * (0.0001/1000)
```

```python
print(f'Estimated cost in USD: {cost: .10f}')
```

```python
def get_embedding(text):
    response = openai.embeddings.create(
        input=text,
        model='text-embedding-ada-002'
    )
    return response.data[0].embedding
```

```python
df['embedding'] = df['text'].apply(lambda x: get_embedding(x))
```

```python
df
```

```python
# storing the embedding Locally
df.to_csv('words-embeddings.csv', index = False)
```

```python
df = pd.read_csv('words-embeddings.csv')
```

```python
df['embedding'] = df['embedding'].apply(eval).apply(np.array) #convert the embedding rows from string to array
```

```python
df
```

```python
search_term = 'fox'
search_term_vector = get_embedding(search_term)
print(search_term_vector)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df['similarities'] = df['embedding'].apply(lambda x: cosine_similarity([x], [search_term_vector])[0][0])
df
```

```python
df = df.sort_values('similarities', ascending = False)
df
```
